```python
def populate(year1, month1,day1, year2, month2,day2,delay):
    
    data = []
    
    sleep(delay)
    
    sdate = date(year1, month1, day1)     
    
    edate = date(year2, month2, day2)
    
    delta = edate - sdate  
            
    for i in range(delta.days + 1):   
        
        day = sdate + timedelta(days=i)
        
        fdat = f"{day:%Y%m%d}"
        
        url = 'http://compras.dados.gov.br/compraSemLicitacao/v1/compras_slicitacao.json?dt_publicacao=' + fdat
    
        req = requests.get(url)
               
        data.append(req.text)            
    
    return data
```
